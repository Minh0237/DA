### Step 5. Delete the column 'Unnamed: 0' and 'Id'

```python
del df['Unnamed: 0']
del df['Id']
```

### Step 6. Is there more male or female names in the dataset?

```python
male = (df['Gender'] == 'M').count()
fmale = (df['Gender'] == 'F').count()
if male > fmale:
  print('male')
else:
  print('female')
```

### Step 7. Group the dataset by name and assign to names

```python
df.drop('Year', axis = 1).groupby('Name').sum()
```

### Step 8. How many different names exist in the dataset?

```python
df['Name'].nunique()
```

### Step 9. What is the name with most occurrences?

```python
df['Name'].mode()

df['Name'].value_counts().idxmax()
```

### Step 10. How many different names have the least occurrences?

```python
least_name = df.groupby('Name').Count.sum().sort_values()[0]
(df.groupby('Name').Count.sum() == least_name).sum()
```

### Step 11. What is the median name occurrence?

```python
median_name = df.groupby('Name').Count.sum().median()
(df.groupby('Name').Count.sum() == median_name).sum()
```

### Step 12. What is the standard deviation of names?

```python
names = df.groupby('Name')
names['Name'].count().std()
```
