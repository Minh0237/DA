### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/u.user).

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/u.user', sep = '|')
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation', dropna = False)['age'].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users[users['gender']=='M'].groupby('occupation')['gender'].count()
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
(users.groupby('occupation')['age']).describe()
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
(users.groupby(['occupation', 'gender'])['age']).mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
Mcount = pd.DataFrame((users[users['gender']=='M'].groupby('occupation')['gender']).count())

for ind in users.index:
  print(Mcount.iloc[ind])
```
