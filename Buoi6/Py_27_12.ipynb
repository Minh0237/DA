```python
import numpy as np

A = np.array([[10, 1, 2],
              [9, 7 ,4],
              [0, 2, 1]])

B = np.array([[2, 3, 4],
              [0, 4, 2],
              [3, 2, 1]])

t1 = np.dot(A, B)
print(t1)

t21 = np.dot(B, A)
t21T = t21.T
t1T = t1.T
t2 = np.dot(t21T, t1T)
print(t2)

t3 = np.dot(t1T, t21T)
print(t3)
```

```python
print(np.linalg.matrix_rank(A))
print(np.linalg.det(A))
print(np.trace(A))
print(np.linalg.pinv(A))
```

```python
print(np.concatenate((A, B), axis = 0))
print(np.reshape(A, -1))
```
